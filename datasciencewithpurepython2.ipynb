```python
#Data Collection from some random instagram users
with open(r"C:\Users\mdosa\OneDrive\Desktop\profiles.txt",encoding = "utf-8") as f:
    data = f.read()
```

```python
#Displaying the data
data
```

```python
#Making the chunks
chunks = data.split("\n\n")
chunks = [chunk for chunk in chunks if len(chunk)>3]
```

```python
#Making function for Parsing the data
def parse_chunk(chunk):
        chunk = chunk.strip()
        sep_chunk = chunk.split("\n")
        username = sep_chunk[0]
        name = sep_chunk[1]
        no_of_posts = int(sep_chunk[2].split(" post")[0].replace(",",""))
        no_of_followers = int(sep_chunk[3].split(" followers")[0].replace(",",""))
        no_of_following = int(sep_chunk[4].split(" following")[0].replace(",",""))
        if (len(sep_chunk)>5):
            bio = "\n".join(sep_chunk[5:])
        else:
            bio =""
        return {"username" : username,"name" : name,"no_of_posts" : no_of_posts,"no_of_followers" : no_of_followers,"no_of_following" : no_of_following,"bio" : bio}
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(all_chunks)
```

```python
import json
s = json.dumps(all_chunks,indent=4)
print(s)
```

```python
#Code to find Maximum posts
max = 0
for chunk in all_chunks:
    if max<chunk["no_of_posts"]:
        max = chunk["no_of_posts"]
        chunk_with_max_no_of_posts = chunk
print(chunk_with_max_no_of_posts)
    
```

```python
#Code to find Maximum followers
max = 0
for chunk in all_chunks:
    if max<chunk["no_of_followers"]:
        max= chunk["no_of_followers"]
        chunk_with_max_no_of_followers = chunk
print(chunk_with_max_no_of_followers)
        
```

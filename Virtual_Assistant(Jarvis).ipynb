```python
#Importing the required libraries
import speech_recognition as sr
import edge_tts
import playsound
import asyncio
import nest_asyncio
import subprocess
import os
import wikipedia
import random
import webbrowser
from datetime import datetime
nest_asyncio.apply()
VOICE = "en-US-ChristopherNeural"
WAKE_WORD = "jarvis"
```

```python
#Text-to-Speech Engine (Jarvis Speaks)
async def speak_async(text):
    print("Jarvis:",text)
    file = "jarvis.mp3"
    communicate = edge_tts.Communicate(text,VOICE)
    await communicate.save(file)
    playsound.playsound(file)
    os.remove(file)
def speak(text):
    asyncio.run(speak_async(text))
```

```python
#Speech recognition (Jarvis Listens)
def listen():
    rec = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        rec.adjust_for_ambient_noise(source,duration = 0.5)
        audio = rec.listen(source)
    try:
        command = rec.recognize_google(audio)
        print("Hi! Osama Sir, you said me: ",command)
        return command.lower()
    except:
        return ""
```

```python
#Time and Date Services
def tell_time():
    time = datetime.now().strftime("%I:%M %p")
    speak(f"The time is {time}")

def tell_date():
    date = datetime.now().strftime("%d %B %Y")
    speak(f"Today's date is {date}")
```

```python
#Website Automation
def open_websites(command):
    if "google" in command:
        speak("Opening Google....")
        webbrowser.open("https://www.google.com")
    elif "instagram" in command:
        speak("Opening Instagram...")
        webbrowser.open("https://www.instagram.com")
    elif "youtube" in command:
        speak("Opening Youtube...")
        webbrowser.open("https://www.youtube.com")
    elif "linkedin" in command:
        speak("Opening LinkedIn...")
        webbrowser.open("https://www.linkedin.com")
    elif "github" in command:
        speak("Opening Github...")
        webbrowser.open("https://www.github.com")
    elif "amazon" in command:
        speak("Opening Amazon...")
        webbrowser.open("https://www.amazon.com")
    elif "chatgpt" in command:
        speak("Opening Chatgpt...")
        webbrowser.open("https://www.chatgpt.com")
```

```python
#Google search automation
def google_search(command):
    if "search" in command:
        query = command.replace("search","")
        speak(f"Searching {query}")
        webbrowser.open(f"https://www.google.com/search?q={query}")
```

```python
#Application control
def open_apps(command):
    if "notepad" in command:
        speak("Opening notepad...")
        os.startfile("notepad.exe")
    elif "calculator" in command:
        speak("Opening calculator...")
        os.startfile("calc.exe")
    elif "msword" in command:
        speak("Opening MS word...")
        os.startfile("winword.exe")
    elif "powerpoint" in command:
        speak("Opening powerpoint ...")
        os.startfile("powerpnt.exe")
```

```python
#Opening browsers
def open_browsers(command):
    if "brave" in command:
        speak("Opening Brave...")
        os.startfile("brave.exe")
    elif "chrome" in command:
        speak("Opening Google Chrome...")
        os.startfile("chrome.exe")
    elif "edge" in command:
        speak("Opening Microsoft edge...")
        os.startfile("msedge.exe")
```

```python
#Play internet radio
def play_radio():
    speak("Playing online radio")
    webbrowser.open("https://www.internet-radio.com/")
```

```python
#Wikipedia information retrieval
def wiki_search(command):
    if "wikipedia" in command:
        try:
            topic = command.replace("wikipedia","")
            speak("Seaching wikipedia")
            result = wikipedia.summary(topic,sentences = 3)
            speak(result)
        except:
            speak("Sorry, I could not find that information")
```

```python
#Conversational intelligence
def smart_talk(command):
    if "who are you" in command:
        speak("I am Jarvis,Your virtual assistant ready to help you sir always!")
    elif "are you married" in command:
        speak("No Sir! i am single and happy")
    elif "hi" in command:
        speak("Hello Osama! How can i help you?")
    elif "how are you" in command:
        speak("I am fine sir,and hope you are also fine and doing well!")
```

```python
#Entertainment module(Jokes)
def tell_jokes():
    jokes = ["why do programmers like dark mode? Because light attracts bugs.","why was the computer cold? Because it forgot to close windows.","I would tell you a python joke, but you might not get the reference."]
    speak(random.choice(jokes))
```

```python
#Main Jarvis Execution Loop
def execute_command(command):
    if "time" in command:
        tell_time()
    elif "date" in command:
        tell_date()
    elif "open" in command:
        open_websites(command)
    elif "search" in command:
        google_search(command)
    elif "joke" in command:
        tell_jokes()
    elif "play music" in command or "play song" in command:
        play_radio()
    elif "wikipedia" in command:
        wiki_search(command)
    elif "notepad" in command or "calculator" in command or "powerpoint" in command or "msword" in command:
        open_apps(command)
    elif "jokes" in command:
        tell_jokes(command)
    elif "talk" in command:
        smart_talk(command)
    elif "exit" in command or "stop" in command or "bye" in command:
        speak("Goodbye, have a nice day")
        return "exit"
```

```python
def run_jarvis():
    speak("Hello osama sir. I am Jarvis. hope you are fine and doing well. System is online.")
    while True:
        heard = listen()
        if heard is None:
            continue
        heard = heard.lower()
        if WAKE_WORD in heard:
            command = heard.replace(WAKE_WORD,"").strip()
            speak("Yes osama sir, i am available for you")
            result = execute_command(command)
            if result == "exit":
                break
                speak("Done with the task , zufisha maam")
run_jarvis()
```
